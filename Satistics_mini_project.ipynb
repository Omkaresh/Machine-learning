```python
#3.	What is the distribution of the winning percentage for all teams?
#Which teams are in the top 5 in terms of winning percentage? 
#(Winning percentage= (GamesWon / GamesPlayed)*100)
df1["GamesWon"]=df1["GamesWon"].astype(int)
df1.GamesPlayed=df1.GamesPlayed.astype(int)
df1["Win_Perct"]= (df1["GamesWon"]/df1.GamesPlayed)*100
df1.Win_Perct
print("Top 5 teams in Win Percentage are \n",)
df1.sort_values(by="Win_Perct",ascending=False)[["Team","Win_Perct"]][:5]

```

```python
#4.	Is there a significant difference in the winning percentage for teams
#which have attained a best position between 1-3 and
#those teams which have had a best position between 4-7.
df1.BestPosition=df1.BestPosition.astype(int)

```

```python
m1=np.array(df1[(df1.BestPosition >=1) & (df1.BestPosition <=3)]["Win_Perct"])
m2=np.array(df1[(df1.BestPosition >=4) & (df1.BestPosition <=7)]["Win_Perct"])
```

```python
t,pval=ttest_ind(m1,m2)
print("t:",t)
print("p-val ",pval)

#Since P value is < 0.05 we can say that there is signiicance difference between the winning percentage
```
